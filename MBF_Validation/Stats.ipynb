```python
from scipy.stats import spearmanr, pearsonr, linregress
import numpy as np

FFR = np.array([0.96, 0.96, 0.97, 0.83, 0.89, 0.94, 0.83, 0.95, 0.87, 0.75, 0.82, 0.75, 0.66, 0.88, 0.81, 0.86, 0.96, 0.9, 0.84, 0.62, 0.6, 0.78, 0.73, 0.88, 0.86, 0.9, 0.84, 0.93, 0.83, 0.9, 0.54, 0.82])
MBFxV = np.array([2312210483, 390642138.8, 487195142.4, 291827877.2, 509058257.4, 783878707.8, 12467536901, 2154825374, 1741860939, 10024240910, 245024544.2, 2902685445, 1525339149, 1525339149, 1525339149, 12627169, 4006239.697, 649922753.7, 4967791102, 2256881799, 10777214174, 100897589.6, 6880517563, 777427783.9, 2547108618, 1589801311, 4282844538, 2948060180, 1504905707, 580419779.9, 3693959838, 761631546.7])
MBFxV /= np.max(MBFxV)
FFR1 = 1- np.array(FFR)
```

```python
print(pearsonr(FFR1, MBFxV))
print(spearmanr(FFR1, MBFxV))
```

```python
import matplotlib.pyplot as plt

plt.scatter(FFR1, MBFxV)

slope, intercept, _, _, _ = linregress(FFR1, MBFxV)
x_fit = np.linspace(min(FFR1), max(FFR1), 100)
y_fit = slope * x_fit + intercept

plt.plot(x_fit, y_fit, 'r--', label=f"Trend line (r={pearsonr(FFR1, MBFxV)[0]:.2f})")
plt.plot(x_fit, x_fit, 'k-', label="x = y line")

plt.xlabel("1 - FFR")
plt.ylabel(r"$\Sigma$MBF x Volume")
plt.legend()

plt.show()
```
