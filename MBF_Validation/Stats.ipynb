```python
import matplotlib.pyplot as plt
FFR = np.array([0.96, 0.96, 0.97, 0.83, 0.95, 0.87, 0.75, 0.82, 0.75, 0.66, 0.88, 0.81, 0.86, 0.96, 0.9, 0.84, 0.62, 0.6, 0.78, 0.73, 0.88, 0.86, 0.9, 0.84, 0.93, 0.83, 0.9, 0.54, 0.82])
MBFxV = np.array([0.143, 0.151, 0.145, 0.269, 0.098, 0.123, 0.11, 0.105, 0.188, 0.094, 0.137, 0.096, 0.134, 0.151, 0.155, 0.164, 0.104, 0.127, 0.131, 0.101, 0.132, 0.122, 0.073, 0.074, 0.121, 0.119, 0.118, 0.099, 0.155])


plt.scatter(FFR, MBFxV)

slope, intercept, _, _, _ = linregress(FFR, MBFxV)
x_fit = np.linspace(min(FFR), max(FFR), 100)
y_fit = slope * x_fit + intercept

plt.plot(x_fit, y_fit, 'r--', label=f"Trend line (r={spearmanr(FFR, MBFxV)[0]:.2f})")
#plt.plot(x_fit, x_fit, 'k-', label="x = y line")

plt.xlabel("FFR")
plt.ylabel(r"$\Sigma$MBF x Volume")
plt.legend()

plt.show()
```
