```python
from scipy.stats import spearmanr, pearsonr, linregress
import numpy as np

FFR = np.array([0.96, 0.96, 0.97, 0.83, 0.89, 0.94, 0.83, 0.95, 0.87, 0.75, 0.82, 0.75, 0.66, 0.88, 0.81, 0.86, 0.96, 0.9, 0.84, 0.62, 0.6, 0.78, 0.73, 0.88, 0.86, 0.9, 0.84, 0.93, 0.83, 0.9, 0.54, 0.82])
MBFxV = np.array([52.2, 22.15, 23.28, 28.4, 40.8, 51.01, 168.75, 41.41, 42.79, 98.6, 14.67, 66.41, 95.5, 111.3, 35.06, 3.74, 2.15, 29.37, 85.15, 44.41, 110.37, 10.37, 77.53, 29.84, 52.2, 31.72, 52.84, 56.42, 39.69, 24.32, 54.78, 31.24])
MBFxV /= np.max(MBFxV)
FFR1 = 1- np.array(FFR)
```

```python
print(pearsonr(FFR1, MBFxV))
print(spearmanr(FFR1, MBFxV))
```

```python
import matplotlib.pyplot as plt

plt.scatter(FFR1, MBFxV)

slope, intercept, _, _, _ = linregress(FFR1, MBFxV)
x_fit = np.linspace(min(FFR1), max(FFR1), 100)
y_fit = slope * x_fit + intercept

plt.plot(x_fit, y_fit, 'r--', label=f"Trend line (r={spearmanr(FFR1, MBFxV)[0]:.2f})")
plt.plot(x_fit, x_fit, 'k-', label="x = y line")

plt.xlabel("1 - FFR")
plt.ylabel(r"$\Sigma$MBF x Volume")
plt.legend()

plt.show()
```
